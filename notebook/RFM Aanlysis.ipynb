```python
import pandas as pd
import datetime
import numpy as np

# https://bit.ly/2Y1cZIQ
import matplotlib.pyplot as plt
import seaborn as sns 
#
sns.set(font='sans-serif')
sns.set_style("whitegrid",{"font.sans-serif":['Microsoft JhengHei']})

# 

# from matplotlib.font_manager import FontProperties
# myfont=FontProperties(fname='./SourceHanSansTC-Bold.ttf',size=14)
# sns.set(font=myfont.get_name())
# sns.set_style("whitegrid",{"font.sans-serif":[myfont.get_name()]})



# # #
purchase_list= pd.read_csv('purchase_list.csv', encoding='utf-8')
purchase_list
recency_label =  ['0-7 day', '8-15 day', '16-22 day', '23-30 day', '31-55 day', '>55 day']
frequency_label =  ['1 freq', '2 freq', '3 freq', '4 freq', '5 freq', '>5 freq']
g = sns.FacetGrid(purchase_list, # 
                  col="recency_cate", # X
                  row="frequency_cate" ,  # Y
                  col_order= recency_label,  # X
                  row_order= frequency_label[::-1], # Y
                  margin_titles=True)
#
g = g.map_dataframe(sns.barplot, # 
                    x= 'gender', # X
                    y ='purchase', # Y ()
                    palette = sns.color_palette("muted"))   #

g = g.set_axis_labels('day','frequency').add_legend()#
```
